```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('dataset/bcdata.sgs.7386.csv', sep=';')
```

```python
# Transforming object to datetime
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y').copy()

#Separeted in years to analysis per 20, 10, 5 years
dfplot = df.copy()
dfplot_20anos = df[df['data'] >= '2000-01-01']
dfplot_10anos = df[df['data'] >= '2010-01-01']
dfplot_5anos = df[df['data'] >= '2015-01-01']
```

```python
plt.figure(figsize=(18,7))
plt.title('Números de Caminhões Vendidos pelas Concessionárias')
sns.lineplot(x=dfplot_20anos['data'], y=dfplot_20anos['valor'], markers=True)
plt.show()
```

```python
plt.figure(figsize=(18,7))
plt.title('Números de Caminhões Vendidos pelas Concessionárias')
sns.lineplot(x=dfplot_10anos['data'], y=dfplot_10anos['valor'], markers=True)
plt.show()
```

```python
plt.figure(figsize=(18,7))
plt.title('Números de Caminhões Vendidos pelas Concessionárias')
sns.lineplot(x=dfplot_5anos['data'], y=dfplot_5anos['valor'], markers=True)
plt.show()
```

```python
dfplot['ano'] = dfplot['data'].dt.year.copy()
dfplot['mes'] = dfplot['data'].dt.month.copy()
dfplot = dfplot[dfplot['mes'] <= 5].copy()
dfplot = dfplot[['ano', 'valor']]
```

```python
dfplot.groupby(by='ano').sum().plot(figsize=(15,7))
```
